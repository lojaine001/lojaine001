```python
import pandas as pd

# Load the data
df = pd.read_csv('weatherHistory1.csv')

# Convert 'Formatted Date' to datetime
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])

# Set 'Formatted Date' as the index
df.set_index('Formatted Date', inplace=True)

# Select the relevant feature
df_temp = df['Temperature (C)']

# Split the data into training and testing sets (80% train, 20% test)
train_size = int(len(df_temp) * 0.8)
train, test = df_temp[:train_size], df_temp[train_size:]

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Plot ACF and PACF for the training data
plot_acf(train)
plot_pacf(train)
plt.show()

from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Define the ARIMA model with p=1, d=1, q=1 as initial guesses
model = ARIMA(train, order=(1, 1, 1))
model_fit = model.fit()

# Forecast
forecast = model_fit.forecast(steps=len(test))

# Plotting
plt.plot(train, label='Train')
plt.plot(test.index, forecast, label='Forecast')
plt.plot(test, label='Test')
plt.legend()
plt.show()

# Print model summary
print(model_fit.summary())

from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Forecast
forecast = model_fit.forecast(steps=len(test))

# Calculate MAE
mae = mean_absolute_error(test, forecast)
print(f'Mean Absolute Error (MAE): {mae}')

# Calculate MSE
mse = mean_squared_error(test, forecast)
print(f'Mean Squared Error (MSE): {mse}')

# Calculate RMSE
rmse = np.sqrt(mse)
print(f'Root Mean Squared Error (RMSE): {rmse}')

# Baseline model: naive forecast (predict the last observed value)
naive_forecast = test.shift(1).dropna()

# Calculate MAE for the naive model
mae_naive = mean_absolute_error(test[1:], naive_forecast)
print(f'Naive Mean Absolute Error (MAE): {mae_naive}')

# Calculate MSE for the naive model
mse_naive = mean_squared_error(test[1:], naive_forecast)
print(f'Naive Mean Squared Error (MSE): {mse_naive}')

# Calculate RMSE for the naive model
rmse_naive = np.sqrt(mse_naive)
print(f'Naive Root Mean Squared Error (RMSE): {rmse_naive}')
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df_temp.values.reshape(-1, 1))

# Create input-output pairs for LSTM
def create_dataset(data, time_steps=1):
    X, Y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:(i + time_steps), 0])
        Y.append(data[i + time_steps, 0])
    return np.array(X), np.array(Y)

time_steps = 12  # Example: Using the past 12 time steps to predict the next time step
X_train, Y_train = create_dataset(scaled_data[:train_size], time_steps)
X_test, Y_test = create_dataset(scaled_data[train_size:], time_steps)

# Reshape input data to be 3D (samples, time steps, features)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense

# Load the data
weather = pd.read_csv("weatherHistory1.csv")

# Ensure that the date column is of type datetime
weather['Formatted Date'] = pd.to_datetime(weather['Formatted Date'])

# Set the date as index
weather.set_index('Formatted Date', inplace=True)

# Select the target feature (e.g., Humidity)
target_feature = 'Humidity'
data = weather[target_feature].values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data)

# Define a function to create input-output sequences for time series forecasting
def create_sequences(data, seq_length):
    X, Y = [], []
    for i in range(len(data) - seq_length):
        end_ix = i + seq_length
        seq_x, seq_y = data[i:end_ix], data[end_ix]
        X.append(seq_x)
        Y.append(seq_y)
    return np.array(X), np.array(Y)

# Define sequence length (e.g., number of time steps to use for prediction)
seq_length = 12  # Example: Using the past 12 time steps to predict the next time step

# Create input-output pairs for the GRU model
X, Y = create_sequences(data_scaled, seq_length)

# Split the data into training and testing sets
train_size = int(len(data_scaled) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
Y_train, Y_test = Y[:train_size], Y[train_size:]

# Define the GRU model architecture
model_gru = Sequential([
    GRU(units=50, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dense(units=1)
])

# Compile the model
model_gru.compile(optimizer='adam', loss='mean_squared_error')

# Train the GRU model
history = model_gru.fit(X_train, Y_train, epochs=100, batch_size=32, validation_data=(X_test, Y_test), verbose=1)

# Plot training and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()

# Make predictions
predictions_gru = model_gru.predict(X_test)

# Inverse transform the predictions and actual values
predictions_gru_inv = scaler.inverse_transform(predictions_gru)
Y_test_inv = scaler.inverse_transform(Y_test)

# Plot actual test data and predictions
plt.plot(Y_test_inv, label='Actual Data')
plt.plot(predictions_gru_inv, label='GRU Predictions')
plt.xlabel('Time')
plt.ylabel('Humidity')
plt.title('Actual vs. Predicted Humidity (GRU)')
plt.legend()
plt.show()


from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Inverse transform the predictions and actual values
predictions_gru_inv = scaler.inverse_transform(predictions_gru)
Y_test_inv = scaler.inverse_transform(Y_test)

# Calculate MAE
mae_gru = mean_absolute_error(Y_test_inv, predictions_gru_inv)

# Calculate MSE
mse_gru = mean_squared_error(Y_test_inv, predictions_gru_inv)

# Calculate RMSE
rmse_gru = np.sqrt(mse_gru)

print('Mean Absolute Error (MAE):', mae_gru)
print('Mean Squared Error (MSE):', mse_gru)
print('Root Mean Squared Error (RMSE):', rmse_gru)
```

**SARIMA**

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# Load the data
weather = pd.read_csv("weatherHistory1.csv")

# Ensure the date column is datetime type
weather['Formatted Date'] = pd.to_datetime(weather['Formatted Date'])

# Set the date as index
weather.set_index('Formatted Date', inplace=True)

# Split the data into training and test sets
train_data = weather['Visibility (km)'][:int(0.8*(len(weather)))]
test_data = weather['Visibility (km)'][int(0.8*(len(weather))):]

# SARIMA Model Training
model = SARIMAX(train_data, order=(1, 0, 0), seasonal_order=(1, 1, 1, 12))  # SARIMA(1,0,0)(1,1,1,12)
model_fit = model.fit()

# Make predictions on the test set
predictions = model_fit.forecast(steps=len(test_data))

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(test_data, predictions))
print('RMSE:', rmse)

# Plot the actual test data and predictions
plt.plot(test_data.index, test_data.values, label='Actual Data')
plt.plot(test_data.index, predictions, color='red', label='Predictions')
plt.xlabel('Date')
plt.ylabel('Visibility (km)')
plt.title('SARIMA Predictions vs. Actual Data')
plt.legend()
plt.show()

# Print RMSE
print('RMSE:', rmse)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Les valeurs de RMSE pour chaque modèle et chaque variable (incluant Naive et sans SES)
models = ['ARIMA', 'Holt-Winters', 'GRU', 'LSTM', 'SARIMA', 'Naive']
temperature_rmse = [8.993423957368316, 35.48, 1.087811315629239, 1.0678255355165913, np.nan, 1.366015835924053]
humidity_rmse = [0.1912034922129761, 0.19, 0.051506640826181165, 0.05132015136526467, 1.6186213098425182, 0.0592198284084283]
visibility_rmse = [4.884431930108511, 19.70, 2.145538976926974, np.nan, np.nan, 2.167450624032763]

# Configuration des plots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))
bar_width = 0.5
index = np.arange(len(models))

def add_value_labels(ax, rects):
    """ Ajoute des labels de valeurs au sommet des barres de diagramme """
    for rect in rects:
        height = rect.get_height()
        if not np.isnan(height):  # Vérifie que la valeur n'est pas NaN
            ax.annotate(f'{height:.2f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points verticaux vers le haut
                        textcoords="offset points",
                        ha='center', va='bottom')
        else:
            ax.annotate('NaN',  # Texte pour NaN
                        xy=(rect.get_x() + rect.get_width() / 2, 0),  # Position en bas
                        xytext=(0, 3),  # 3 points verticaux vers le haut
                        textcoords="offset points",
                        ha='center', va='bottom', color='red')

# Plot pour la Température
temp_bars = axs[0].bar(index, temperature_rmse, bar_width, color='skyblue', label='Temperature')
axs[0].set_title('ACCURACY for Temperature')
axs[0].set_xlabel('Models')
axs[0].set_ylabel('RMSE')
axs[0].set_xticks(index)
axs[0].set_xticklabels(models)
axs[0].legend()
add_value_labels(axs[0], temp_bars)

# Plot pour l'Humidité
hum_bars = axs[1].bar(index, humidity_rmse, bar_width, color='lightgreen', label='Humidity')
axs[1].set_title('ACCURACY for Humidity')
axs[1].set_xlabel('Models')
axs[1].set_ylabel('RMSE')
axs[1].set_xticks(index)
axs[1].set_xticklabels(models)
axs[1].legend()
add_value_labels(axs[1], hum_bars)

# Plot pour la Visibilité
vis_bars = axs[2].bar(index, visibility_rmse, bar_width, color='lightcoral', label='Visibility')
axs[2].set_title('ACCURACY for Visibility')
axs[2].set_xlabel('Models')
axs[2].set_ylabel('RMSE')
axs[2].set_xticks(index)
axs[2].set_xticklabels(models)
axs[2].legend()
add_value_labels(axs[2], vis_bars)

# Ajustement de l'espacement entre les sous-plots
plt.tight_layout()
plt.show()
```
